```python
def cleaning_score(score):
    parts = score.split(':')
    if len(parts) == 2:
        first, second = parts
        second = second.lstrip('0')
        if not second:
            return f"{first}:0"
        else:
            return f"{first}:{second}"
    else:
        return score
    
def convert_numeric(minute):
    if '+' in minute:
        parts = minute.split('+')
        numeric_value = int(parts[0]) + int(parts[1])
    else:
        numeric_value = int(minute)
    return numeric_value

def address_season(season):
    season = season.replace("Dec", "12").replace("-", "/")
    return season

def address_competition(competition):
    competition = competition.replace("Champions League", "UEFA Champions League")
    competition = competition.replace("Troph�e des Champions", "Trophée des Champions")
    return competition

def address_matchday(matchday):
    matchday = matchday.replace("final", "Final").replace("last 16", "Round of 16")
    return matchday

def address_date(date):
    date = date.replace("-", "/")
    return date

def format_date(fecha):
    partes = fecha.split('/')
    partes = [parte.zfill(2) if len(parte) == 1 else parte for parte in partes]
    return '/'.join(partes)

def address_type(Type):
    if Type in ['Solo run', 'Tap-in', 'Deflected shot on goal', 'Counter attack goal', 'Long distance kick']:
        return 'Left-footed shot'
    elif Type == 'Penalty rebound':
        return 'Penalty'
    else:
        return Type

df['Result'] = df['Result'].apply(cleaning_score)
df['At_score'] = df['At_score'].apply(cleaning_score)
df['Minute'] = df['Minute'].apply(convert_numeric)
df['Season'] = df['Season'].apply(address_season)
df['Competition'] = df['Competition'].apply(address_competition)
df['Competition'] = df['Competition'].str.strip().replace('UEFA UEFA Champions League', 'UEFA Champions League')
df['Matchday'] = df['Matchday'].apply(address_matchday)
df['Date'] = df['Date'].apply(address_date)
df['Date'] = df['Date'].apply(format_date)
df['Type'] = df['Type'].apply(address_type)

df['Playing_Position'] = df['Playing_Position'].str.strip()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y', errors='coerce')
```
